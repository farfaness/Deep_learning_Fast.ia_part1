## Creation of a ImageDataBunch object with the data block API

To put this in a `DataBunch` while using the [data block API](https://docs.fast.ai/data_block.html), we then need to using `ImageList` (and not `ImageDataBunch`). This will make sure the model created has the proper loss function to deal with the multiple classes.

```python
# DATA AUGMENTATION

tfms = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)

# Flip vert :
# Transforms by default will flip randomly the images horizontaly but not vertically
# so here we put flip_vert: true because it is satellite images 
#(also for molecules ... everything that have no up and down)
#(now flip vertically, and each 90 degres rotations)

# Perspective warping : 
# difference between looking at someone from below or above, the shape changes, 
# à inclure pour image de chien par exemple mais pas pour images de satellite
```

We use parentheses around the data block pipeline below, so that we can use a multiline statement without needing to add '\\'.

With ImageList we are going to create by ourselves a Imagedatabunch object!! For that we need : where the images comes from, where the labels comme from, and how to create a validation set.

```python
np.random.seed(42)
src = (ImageList.from_csv(path, 'train_v2.csv', folder='train-jpg', suffix='.jpg')
       .split_by_rand_pct(0.2)
       .label_from_df(label_delim=' '))

# How to split in train/valid? -> use random 20%
# How to label? -> depending on the dataframe
```

```python
# Fonction to transform them in Databunch Image object (create the dataloader and databunch)!
data = (src.transform(tfms, size=128)
        .databunch().normalize(imagenet_stats))

# Data augmentation? -> use tfms with a size of 128
# Create a databunch
# Normalize the databunch
```

`show_batch` still works, and show us the different labels separated by `;`.

```python
data.show_batch(rows=3, figsize=(12,9))
```

To create a `Learner` we use the same function as in lesson 1. Our base architecture is resnet50 again, but the metrics are a little bit differeent: we use `accuracy_thresh` instead of `accuracy`. In lesson 1, we determined the predicition for a given class by picking the final activation that was the biggest, but here, each activation can be 0. or 1. `accuracy_thresh` selects the ones that are above a certain threshold (0.5 by default) and compares them to the ground truth.

As for Fbeta, it's the metric that was used by Kaggle on this competition. See [here](https://en.wikipedia.org/wiki/F1_score) for more details.

```python
arch = models.resnet50
```

```python
acc_02 = partial(accuracy_thresh, thresh=0.2)
# On crée une partial function application (c'est comme faire un def et créer une nouvelle fonction): metric accuracy avec threshold 0.2
# Avec multilabel on peut pas faire de argmax pour déterminer la class choisie par le model pour calculer l'accuracy
# A la place : on va utiliser data.c (nbr de classes de target), on compare chacune des proba des classes à un threshold
# tout ce qui est plus important que ce threshold, on va dire que l'image à ce feature
# Pour ce type de dataset un threshold de 0.2 fonctionne bien

f_score = partial(fbeta, thresh=0.2)

learn = cnn_learner(data, arch, metrics=[acc_02, f_score])

# Different metrics printed out here : accuracy, et f_score (metrics used in this kaggle to judge)
# f_score : combine false positiv and negativ in a single number
```

We use the LR Finder to pick a good learning rate.

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

Forme classique que l'on obtient souvent avant de unfreeze

Then we can fit the head of our network (derniers layers)



```python
lr = 0.01
```

```python
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-rn50')
```

...And fine-tune the whole model:

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

Après unfreeze, on obtient souvent une forme bien différente, qui ressemble souvent à ça. Pour slice, pour premières couches prendre le moment où ça se met à augmenter d'un coup, et faire 1e-1. (du coup ici on pass de 1e-4 à 1e-5).

Et pour les dernières couches, on prend la valeur que l'on avait en unfrozen (ici 0.01), et on divise par 5 ou par 10 cette valeur (pour que soit encore un peu plus lent).

-> Discriminative learning rate

```python
0.01/5 #lr/5
```

```python
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

```python
learn.save('stage-2-rn50')
```

On veut être dans les 10% meilleurs de Kaggle, c'est pas encore ça, il faut encore améliorer notre model !!!

# More tuning to improve kaggle classment : Progressive Resizing tweak

i.e. transfert learning on the same dataset but with a bigger size 
when using this tweak, it is best to start with a minimal size of 64*64, and inreases it

```python
# On avait prit une taille de 128 avant pour aller vite, on passe à 256
# On va faire du transfert learning et reutiliser notre model qui est bon à détecter des images stallite de taille 128*128
# et le fine-tuner pour qu'il puisse reconnaitre des images de 256*256

# En plus comme on change la taille de notre dataset, c'est comme si on avait un nouveau dataset
# du coup on perd les risques d'overfitting

# on garde le same learning, mais avec un new databunch (on garde même source pour les images, on change juste la taille)
data = (src.transform(tfms, size=256)
        .databunch().normalize(imagenet_stats))

# On garde le même learner que l'on vient d'entrainer, 
#on va juste remplacer les data dans le learner avec le nouveau databunch que l'on vient de créer
learn.data = data
data.train_ds[0][0].shape
```

```python
learn.freeze() # On freeze à nouveau, pour entrainer juste les quelques derniers layers
```

```python
learn.lr_find()
learn.recorder.plot()
```

Comme on a un bon model pré-entrainé sur une taille plus petite, on voit que la courbe ne ressemble pas à celle d'un freezed model. 

```python
lr=1e-2/2; lr
```

```python
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-256-rn50')
```

On déjà ici dans les top 10% de la compétition Kaggle!!!

Mais on peut comme même unfreeze et entrainer encore un peu notre modèle. 

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(5, slice(1e-5, lr/5))
```

On est maintenant dans le top 20 de la compétition!!

```python
learn.recorder.plot_losses()
```

```python
learn.save('stage-2-256-rn50')
```
